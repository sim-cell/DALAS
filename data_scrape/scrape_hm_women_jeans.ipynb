```python
import bs4
import lxml
import pandas as pd
import urllib

from urllib import request
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
url = 'https://www2.hm.com/en_us/women/products/jeans.html'
```

```python
#scrape all links to jeans from main catalog with their price
main_link = 'https://www2.hm.com'
links = []

url = main_link+'/en_us/women/products/jeans.html?sort=stock&image-size=small&image=model&offset=0&page-size=200'
req = requests.get(url,headers=headers)
page = bs4.BeautifulSoup(req.content, "lxml")
items = page.find_all('li',class_='product-item')
for item in items:
    link = item.find('a', class_='link')
    if link:
        href = link.get('href')
        links.append(href)
links
```

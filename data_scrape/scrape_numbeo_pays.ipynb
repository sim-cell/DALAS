```python
import bs4
import lxml
import pandas as pd
import urllib

from urllib import request
```

```python
# pays
url = 'https://www.numbeo.com/pollution/'
req = request.urlopen(url).read()
page = bs4.BeautifulSoup(req, "lxml")
pays_list = page.find_all('option')

pays_list = [p.text for p in pays_list if p.get('value')]
pays_list
```

```python
df = pd.DataFrame()
df["Country"] = pays_list
df
```

Scraping the link for each country one by one

```python
def find_table(url):
    req = request.urlopen(url).read()
    page = bs4.BeautifulSoup(req, "lxml")
    table = page.find('table',{"class":"table_builder_with_value_explanation data_wide_table"})
    return table
```

```python
url_pays = 'https://www.numbeo.com/pollution/country_result.jsp?country='#Australia
#init dataframe with the first country
list_data = []
url = url = url_pays+pays_list[0].replace(" ","+")
columns = ['Country']
data = [pays_list[0]]

table = find_table(url)
columns += [td.text.strip() for td in table.find_all('td',class_='columnWithName')]
print(columns)
data +=  [float(td.text.strip()) for td in table.find_all('td',class_="indexValueTd" )]
list_data.append(data)
```

```python
for pays in pays_list[1:]:
    url = url_pays+pays.replace(" ","+")
    table = find_table(url)
    data = [pays]
    data +=  [td.text.strip() for td in table.find_all('td',class_="indexValueTd" )]
    list_data.append(data)
```

```python
df_pollution = pd.DataFrame(list_data,columns=columns)
df_pollution
```

```python
df_pollution.to_csv('../data/data_pollution_init.csv', index=False)
```

```python
#cleanup
df_pollution_clean = df_pollution.copy()
df_pollution_clean.replace('?',pd.NA,inplace=True) #changing ? to NA
df_pollution_clean.dropna(axis=0,inplace=True)
df_pollution_clean.reset_index(drop=True, inplace=True)
df_pollution_clean
```

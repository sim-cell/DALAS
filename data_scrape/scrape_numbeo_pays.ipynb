```python
import bs4
import lxml
import pandas as pd
import urllib

from urllib import request
```

```python
# pays
url = 'https://www.numbeo.com/pollution/'
req = request.urlopen(url).read()
page = bs4.BeautifulSoup(req, "lxml")
pays_list = page.find_all('option')

pays_list = [p.text for p in pays_list if p.get('value')]
pays_list
```

```python
df = pd.DataFrame()
df["Country"] = pays_list
df
```

Scraping the link for each country one by one
